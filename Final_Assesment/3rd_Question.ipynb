# Q3 : Plot average BoxOffice Collection by Genre for combined dataset
What is your inference from this?

How does average change when we pick only Bollywood movies?

What is your inference now?


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
indian_movies = pd.read_excel('cleaned_movies_data/cleaned_indian_movies.xlsx')
bollywood_movies = pd.read_excel('cleaned_movies_data/cleaned_bollywood_movies.xlsx')
tollywood_movies = pd.read_excel('cleaned_movies_data/cleaned_tollywood_movies.xlsx')

# Combine all datasets
combined_movies = pd.concat([indian_movies, bollywood_movies, tollywood_movies], ignore_index=True)

# Calculate average BoxOffice collection by Genre for the combined dataset
avg_combined = combined_movies.groupby('Genre')['BoxOffice (Crores)'].mean().sort_values(ascending=False)

# Calculate average BoxOffice collection by Genre for Bollywood movies
avg_bollywood = bollywood_movies.groupby('Genre')['BoxOffice (Crores)'].mean().sort_values(ascending=False)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Combined dataset plot
axes[0].barh(avg_combined.index, avg_combined.values, color='teal')
axes[0].set_title('Average BoxOffice Collection by Genre (All Movies)')
axes[0].set_xlabel('Average BoxOffice Collection (Crores)')
axes[0].set_ylabel('Genre')

# Bollywood dataset plot
axes[1].barh(avg_bollywood.index, avg_bollywood.values, color='orange')
axes[1].set_title('Average BoxOffice Collection by Genre (Bollywood)')
axes[1].set_xlabel('Average BoxOffice Collection (Crores)')
axes[1].set_ylabel('Genre')

plt.tight_layout()
plt.savefig('Q3_output.png')
plt.show()
```

## Inference 
1. Similar to the combined dataset, Action and Drama genres likely dominate in terms of average collections.
2. However, Romance may also show a significant average collection, reflecting Bollywood's historical strength in producing romantic films that resonate with audiences.
3. The average collections for genres like Comedy may also be notable, indicating that Bollywood successfully capitalizes on humor and entertainment.
