### 1. Analysis: Relationship Between Budget and BoxOffice Collections

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the movie data from Excel files
indian_movies = pd.read_excel('cleaned_movies_data/cleaned_indian_movies.xlsx')

# Scatter plot to visualize relationship
plt.figure(figsize=(8,6))
plt.scatter(indian_movies['Budget (Crores)'], indian_movies['BoxOffice (Crores)'], alpha=0.5)
plt.title('Budget vs BoxOffice Collection')
plt.xlabel('Budget (Crores)')
plt.ylabel('BoxOffice Collection (Crores)')
plt.grid(True)
plt.tight_layout()
plt.savefig('Q4_output1.png')
plt.show()

# Correlation calculation
corr = indian_movies[['Budget (Crores)', 'BoxOffice (Crores)']].corr().iloc[0,1]
print(f"Correlation between Budget and BoxOffice Collection: {corr:.2f}")
```

### 2. Analysis: Year-wise Trend of Number of Movies Released

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the movie data from Excel files
indian_movies = pd.read_excel('cleaned_movies_data/cleaned_indian_movies.xlsx')

# Count movies released per year
yearly_counts = indian_movies.groupby('ReleaseYear').size()

# Plotting
plt.figure(figsize=(10,6))
yearly_counts.plot(kind='line', marker='o')
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.tight_layout()
plt.savefig('Q4_output2.png')
plt.show()
```

### 3. Bonus Analysis: Which Language Has Made the Most Movies?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the movie data from Excel files
indian_movies = pd.read_excel('cleaned_movies_data/cleaned_indian_movies.xlsx')

# Count number of movies per language or industry
language_counts = indian_movies['Language'].fillna('Unknown').value_counts()

# Print the counts
print("Number of movies made by each language:")
print(language_counts)

# Plot as a simple bar chart
language_counts.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('Number of Movies by Language')
plt.xlabel('Language')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Q4_output3.png')
plt.show()
```
